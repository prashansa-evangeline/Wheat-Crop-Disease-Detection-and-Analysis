```python
import os
import pandas as pd
import numpy as np

def load_ground_truths(annotation_path):
    # Assuming the format is one where each line corresponds to a different object
    # in the format: class x_center y_center width height
    data = []
    for file in os.listdir(annotation_path):
        if file.endswith(".txt") and not file.startswith("classes"):
            file_path = os.path.join(annotation_path, file)
            with open(file_path, 'r') as f:
                for line in f:
                    class_id = int(line.split()[0])  # Assuming first element is class ID
                    data.append((file, class_id))
    return pd.DataFrame(data, columns=['file', 'true_label'])

def load_predictions(prediction_path):
    data = []
    for file in os.listdir(prediction_path):
        if file.endswith(".txt"):
            file_path = os.path.join(prediction_path, file)
            with open(file_path, 'r') as f:
                for line in f:
                    parts = line.strip().split()
                    class_id = int(parts[0])
                    confidence = float(parts[-1])  # Ensure this is the correct index for confidence
                    if confidence >= 0.4:  # Adjust the threshold as needed
                        data.append((file.replace('.txt', ''), class_id))
    return pd.DataFrame(data, columns=['file', 'predicted_label'])

# Correct paths
ground_truth_path = '/content/drive/MyDrive/FHB/valid/labels'  # Ensure this is where your ground truth labels are
prediction_path = '/content/yolov5/runs/detect/exp/labels'

y_true_df = load_ground_truths(ground_truth_path)
y_pred_df = load_predictions(prediction_path)

# Merge and prepare for analysis
results_df = pd.merge(y_true_df, y_pred_df, on='file', how='left').fillna(-1)  # -1 for missing predictions

# Now you have a DataFrame with true labels and predicted labels
y_true = results_df['true_label'].tolist()
y_pred = results_df['predicted_label'].tolist()

# Now you could generate a classification report using sklearn
```

```python
unique_classes = set(y_true).union(set(y_pred))
print("Unique classes in the dataset:", unique_classes)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_true, y_pred, target_names=['diseasedPart', 'wheatHead', "undetected"]))
```

```python
from sklearn.metrics import precision_recall_fscore_support



# Calculate precision, recall, and F1 score
precision, recall, fscore, _ = precision_recall_fscore_support(y_true, y_pred, average='weighted')

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {fscore:.2f}")
```
